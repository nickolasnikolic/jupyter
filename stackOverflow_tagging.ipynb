```python
from stackapi import StackAPI as stack
import json
import time
from vowpalwabbit import pyvw as wabbit

model = wabbit.vw(quiet=True)

s = stack('stackoverflow')
r = s.fetch('questions')
```

```python
for item in r.items():
  title = ''
  if item[0] == 'items':
    for i in list(item[1]):
      tags = ''
      for tag in list(i['tags']):
        tags += tag + ' '
      title = i['title']
      if title != '' and tags != '':
        print(title)
        print(tags)
        model.learn( title + ' ' + tags )
      predicted = model.predict('my butt hurts' + ' ' + 'butt hurts')
      print(predicted)
```
