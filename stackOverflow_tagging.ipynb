```python
from stackapi import StackAPI as stack
import json
import time
from vowpalwabbit import pyvw as wabbit

model = wabbit.vw(quiet=True)

s = stack('stackoverflow')
r = s.fetch('questions')
print(r)
```

```python
for result in list(r.items):
  print(result)
  for tag in list(result.tags):
    model.learn( result.title, tag )

predicted = model.predict('my butt hurts', 'butt')
print(predicted)
```
